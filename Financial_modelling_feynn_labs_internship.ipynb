```python
#FINANCIAL MODELLING

import pandas as pd
df1=pd.read_csv("supermarket_sales - Sheet1.csv")
```

```python
df1.head()
```

```python
#using scikit-learn
#Import LabelEncoder using sklearn
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1["Branch"] = le.fit_transform(df1["Branch"])
df1["City"] = le.fit_transform(df1["City"])
df1["Customer type"] = le.fit_transform(df1["Customer type"])
df1["Gender"] = le.fit_transform(df1["Gender"])
```

```python
X = df1.drop("gross income",axis = 1 )
X=df1.drop("Invoice ID", axis=1)
y = df1["gross income"]
```

```python
df1['Product line'].unique()
```

```python
df1=df1.drop("Invoice ID", axis=1)
df1=df1.drop("Date", axis=1)
df1=df1.drop("Time", axis=1)
df1=df1.drop("Payment", axis=1)
```

```python
df_Health_and_beauty=df1.where(df1['Product line']=='Health and beauty')
df_Health_and_beauty=df_Health_and_beauty.dropna()
```

```python
df_Health_and_beauty.reset_index()
df_Health_and_beauty.head()
```

```python
df_Electronic_accessories=df1.where(df1['Product line']=='Electronic accessories')
df_Electronic_accessories=df_Electronic_accessories.dropna()
```

```python
df_Electronic_accessories.reset_index()
df_Electronic_accessories.head()
```

```python
df_Home_and_lifestyle=df1.where(df1['Product line']=='Home and lifestyle')
df_Home_and_lifestyle=df_Home_and_lifestyle.dropna()
df_Home_and_lifestyle.reset_index()
```

```python
df_Sports_and_travel=df1.where(df1['Product line']=='Sports and travel')
df_Sports_and_travel=df_Sports_and_travel.dropna()
df_Sports_and_travel.reset_index()
```

```python
df_Food_and_beverages=df1.where(df1['Product line']=='Food and beverages')
df_Food_and_beverages=df_Food_and_beverages.dropna()
df_Food_and_beverages.reset_index()
```

```python
df_Fashion_and_accessories=df1.where(df1['Product line']== 'Fashion accessories')
df_Fashion_and_accessories=df_Fashion_and_accessories.dropna()
df_Fashion_and_accessories.reset_index()
```

```python
X_HB = df_Health_and_beauty.drop("Product line",axis = 1 )
X_EC = df_Electronic_accessories.drop("Product line",axis = 1 )
X_HL = df_Home_and_lifestyle.drop("Product line",axis = 1 )
X_ST = df_Sports_and_travel.drop("Product line",axis = 1 )
X_FB = df_Food_and_beverages.drop("Product line",axis = 1 )
X_FA = df_Fashion_and_accessories.drop("Product line",axis = 1 )
```

```python
y_HB = df_Health_and_beauty["gross income"]
y_EC = df_Electronic_accessories["gross income"]
y_HL = df_Home_and_lifestyle["gross income"]
y_ST = df_Sports_and_travel["gross income"]
y_FB = df_Food_and_beverages["gross income"]
y_FA = df_Fashion_and_accessories["gross income"]
```

```python
X_HB
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train_HB,X_test_HB, y_train_HB,y_test_HB = train_test_split(X_HB,y_HB,test_size=0.25,random_state = 15062001 )
```

```python
regr_HB=LinearRegression()
regr_HB.fit(X_train_HB,y_train_HB)
```

```python
print(regr_HB.score(X_test_HB, y_test_HB))
```

```python
X_train_EC,X_test_EC, y_train_EC,y_test_EC = train_test_split(X_EC,y_EC,test_size=0.25,random_state = 15062001 )
```

```python
regr_EC=LinearRegression()
regr_EC.fit(X_train_EC,y_train_EC)
```

```python
print(regr_EC.score(X_test_EC, y_test_EC))
```
